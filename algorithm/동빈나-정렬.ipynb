### 선택정렬

- 처리되지 않은 데이터 중 가장 작은 데이터를 선택한 후 맨 앞에 있는 데이터와 바꾸는 것을 반복한다


```python
#예제코드-가장 작은 수 부터 오름차순 정렬

arr=[7,5,9,0,3,1,6,2,4,8]

for i in range(len(arr)):
    min_index=i #min_index은 i번째 값으로 초기화
    for k in range(i+1,len(arr)): #처리되지 않은 데이터 중 가장 작은 데이터 고르기
        if arr[min_index]>arr[k]:
            min_index=k #작은 for문 수행이 끝나면 가장 작은 원소의 min_index가 큰 for문의 min_index에 들어가게됨
            
    #이때까지 가장 작은 값의 인덱스와 현재 min_index값에 해당하는 원소의 값을 스와핑
    arr[i],arr[min_index]=arr[min_index],arr[i]
        
```

```python
print(arr)
```

### 삽입정렬
- 처리되지 않은 데이터를 하나씩 골라 적절한 위치에 삽입
- 구현 난이도는 비교적 높지만 선택정렬보다 효율적으로 동작하는 편
- 현재 리스트의 데이터가 거의 정렬되어있는 상태라면 매우 빠르게 동작함
- 삽입정렬의 시간복잡도는 O(N^2)이지만 위 경우는 O(N)의 시간복잡도를 가짐


```python

arr=[7,5,9,0,3,1,6,2,4,8]

for i in range(1,len(arr)):
    for j in range(i,0,-1):
        #j: 삽입 대상인 원소의 위치
        #앞에 위치한 원소보다 값이 작으면 스와프
        if arr[j]<arr[j-1]:
            arr[j-1],arr[j]=arr[j],arr[j-1]
        else:#크거나 같으면 그대로
            break
            
print(arr)
```
